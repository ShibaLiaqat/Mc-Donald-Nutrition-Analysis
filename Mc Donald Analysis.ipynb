## Which variables have the highest correlation? 

```python
cor=McD.corr()
plt.figure(figsize=(15, 15))
sns.heatmap(cor,annot = True, vmin=-1, vmax=1, center= 0)
```

As per the heatmap above and the correlation matrix, the variable with the highest correlation is
Top 5 highest correlation is between -

Calories from Fat - Total Fat : 0.99

Total Fat Daily Value.- Total Fat : 0.99

Calories-Total Fat Daily Value : 0.99

Saturated.Fat-Saturated.Fat Daily Value : 0.99

Cholestrol- Cholesterol Daily Value : 0.99

## Which item contributes maximum to the Sodium intake?

```python
McD.groupby('Item')['Sodium'].sum().sort_values(ascending = False)
```

## Which 4 food items contains the most amount of Saturated Fat?

```python
McD.groupby('Item').sum()['Saturated Fat'].sort_values(ascending = False)
```

##  Which category contributes to the maximum % of Cholesterol in a diet (% daily value)?

```python
McD.groupby('Category').sum()['Cholesterol (% Daily Value)'].sort_values(ascending = False)
```
